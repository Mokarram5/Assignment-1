#### Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution..

#### Answer:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'https://storage.googleapis.com/flight-data8273/flight_price.csv'

flight_data = pd.read_csv(file_path)

print(flight_data.head())

plt.figure(figsize=(10, 6))
sns.histplot(flight_data['Price'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

#### Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

#### Answer:

```python
import pandas as pd

file_path = 'https://storage.googleapis.com/flight-data8273/flight_price.csv'

flight_data = pd.read_csv(file_path)

# Find the minimum and maximum prices
min_price = flight_data['Price'].min()
max_price = flight_data['Price'].max()

# Calculate the range
price_range = max_price - min_price

# Print the results
print(f"The minimum price is: {min_price}")
print(f"The maximum price is: {max_price}")
print(f"The range of prices is: {price_range}")
```

#### Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines..

#### Answer:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'https://storage.googleapis.com/flight-data8273/flight_price.csv'

flight_data = pd.read_csv(file_path)

# Create a boxplot to compare prices by airline
plt.figure(figsize=(14, 8))
sns.boxplot(x='Airline', y='Price', data=flight_data, palette='viridis')
plt.title('Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

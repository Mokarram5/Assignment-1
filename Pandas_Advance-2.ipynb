#### Q6. What do you understand about the windows function in pandas and list the types of windows functions?

#### Answer:

A window function (also known as a windowing or rolling function) is a feature that allows us to perform calculations on a specific subset of data in a DataFrame or Series.

The key idea behind window functions is to create a rolling view over the data, and then apply a function to that rolling view. we can customize the window's size, shape, and behavior to suit your specific analytical needs. Some common window functions include:

1. **rolling()**: This function creates a rolling view over a Series or DataFrame, allowing you to perform rolling calculations. You can specify the window size using the `window` parameter and apply various aggregation functions like mean, sum, min, max, etc.

```python

df['rolling_mean'] = df['column'].rolling(window=3).mean()
```

2. **expanding()**: This function creates an expanding view over the data, which means that the window size increases with each step. It's useful for computing cumulative statistics or other expanding calculations.

```python

df['expanding_sum'] = df['column'].expanding().sum()
```

3. **ewm() (Exponentially Weighted Moving Average)**: This function is used for calculating exponentially weighted statistics. It assigns exponentially decreasing weights to the data points in the window. You can specify parameters like `span`, `halflife`, or `alpha` to control the weighting.

```python

df['ewma'] = df['column'].ewm(span=3).mean()
```

4. **rolling_apply()**: This function allows you to apply a custom function to a rolling window. You can define your own function and pass it as an argument to this function.

```python

def custom_function(window):
    # Your custom calculation here
    return result

df['custom_result'] = df['column'].rolling(window=3).apply(custom_function)
```

5. **shift()**: While not a typical window function, the `shift()` function allows you to access values from previous or future positions in the Series or DataFrame. It can be useful in various time series calculations.

```python

df['value_diff'] = df['column'] - df['column'].shift(1)
```

#### Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and calculates the difference between them in days, hours, and minutes using Pandas time delta. The program should prompt the user to enter the dates and display the result.

#### Answer:

````python
import pandas as pd

def calculate_time_difference(start_date, end_date):
    start_datetime = pd.to_datetime(start_date)
    end_datetime = pd.to_datetime(end_date)
    time_difference = end_datetime - start_datetime
    return time_difference
````

````python
def display_time_difference(time_difference):
    days = time_difference.days
    hours, remainder = divmod(time_difference.seconds, 3600)
    minutes, _ = divmod(remainder, 60)
    return days, hours, minutes
````

````python
start_date = input("Enter the start date (YYYY-MM-DD): ")
end_date = input("Enter the end date (YYYY-MM-DD): ")

time_difference = calculate_time_difference(start_date, end_date)

days, hours, minutes = display_time_difference(time_difference)
print(f"Time difference: {days} days, {hours} hours, {minutes} minutes")
````

#### Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified column to a categorical data type. The program should prompt the user to enter the file path, column name, and category order, and then display the sorted data.

#### Answer:

````python
import pandas as pd
def convert_column_to_categorical(file_path,column_name, category_order):
    try:
        
        df = pd.read_csv(file_path)
        
        df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)
        
        
        sorted_df = df.sort_values(by=column_name)
        
        return sorted_df
        
    except FileNotFoundError:
        print(f"File not found: {file_path}")
    except KeyError:
        print(f"Column not found: {column_name}")

file_path = input("Enter the CSV file path: ")
column_name = input("Enter the column name to convert to categorical: ")
category_order = input("Enter the category order (comma-separated): ").split(",")

sorted_data = convert_column_to_categorical(file_path, column_name, category_order)

if sorted_data is not None:
    print("\nSorted Data:")
    print(sorted_data)
````

#### Q10. Write a Python program that reads a CSV file containing sales data for different products and visualizes the data using a stacked bar chart to show the sales of each product category over time. Theeprogram should prompt the user to enter the file path and display the chart.

````python
import pandas as pd
import matplotlib.pyplot as plt

def visualize_sales_data(file_path):
    try:

        df = pd.read_csv(file_path)
        
        sales_data = df.groupby(['Month', 'ProductSales'])
        
        sales_data.plot(kind='bar', stacked=True, figsize=(10, 6))
        
        plt.xlabel('Month')
        plt.ylabel('ProductSales')
        plt.title('Stacked Bar Chart of Sales by Product Category Over Time')
        
        plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
        plt.show()
    except FileNotFoundError:
        print(f"File not found: {file_path}")

file_path = input("Enter the CSV file path: ")
visualize_sales_data(file_path)
````
